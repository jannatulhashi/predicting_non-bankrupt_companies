```python
# Dependencies and Setup
import os
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
# Store filepath in a variable
cyber_security_breaches_filepath = Path("Resources/cyber_security_breaches.csv")
```

```python
data_breaches_df = pd.read_csv(cyber_security_breaches_filepath)
data_breaches_df.head()
```

```python
data_breaches_df.info()
```

```python
# Drop 'Unnamed: 0' and 'Business_Associate_Involved'
data_breaches_df.drop(columns=['Unnamed: 0'], inplace=True)

data_breaches_df.drop(columns=['Business_Associate_Involved'], inplace=True)
```

```python
data_breaches_df.head()
```

```python
# Convert date columns to datetime
data_breaches_df['Date_of_Breach']= pd.to_datetime(data_breaches_df['Date_of_Breach'], errors = 'coerce')

data_breaches_df['Date_Posted_or_Updated'] =  pd.to_datetime(data_breaches_df['Date_Posted_or_Updated'] , errors = 'coerce')

data_breaches_df['breach_start'] =  pd.to_datetime(data_breaches_df['breach_start'] , errors = 'coerce')

data_breaches_df['breach_end'] = pd.to_datetime(data_breaches_df['breach_end'] , errors = 'coerce')
```

```python
# Rename specified columns
data_breaches_df.rename(columns={
    'Name_of_Covered_Entity': 'Entity_Name',
    'breach_start': 'Breach_Start',
    'breach_end' : 'Breach_End',
    'year' : 'Year',
    'lat' : 'Lat',
    'lon' : 'Lon'
}, inplace=True)

```

```python
data_breaches_df
```

```python
data_breaches_df.dtypes
```

```python
data_breaches_df.isnull().sum()
```

```python
# Fill missing values
data_breaches_df['Date_of_Breach'].fillna('Unknown', inplace = True)
data_breaches_df['Summary'].fillna('Unknown', inplace = True)
data_breaches_df['Breach_End'].fillna('Unknown', inplace = True)
data_breaches_df['Lat'].fillna('Unknown', inplace = True)
data_breaches_df['Lon'].fillna('Unknown', inplace = True)
```

```python
data_breaches_df.isnull().sum()
```

```python
# Calculate the number of duplicate rows in the dataframe
number_of_duplicates = data_breaches_df.duplicated().sum()

# Print the number of duplicate rows
print(f"There are {number_of_duplicates} duplicate rows in the dataframe.")
```
